```python
# 2차 함수

def quadratic_function(x):
    a = 1
    b = -2
    c = -2
    
    return a* x**2 + b*x + c
```

```python
x = np.arange(-5, 5, 0.1)
y = quadratic_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
```

```python
# 3차 함수

def cubic_function(x):
  a = 4
  b = 0
  c = -1
  d = -8

  return a*x**3+b*x**2+c*x+d
```

```python
x = np.arange(-5, 5, 0.1)
y = cubic_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Function')
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
def AND(x1, x2): 
    input = np.array([x1, x2])             #자극
    weights = np.array([0.4,0.4])          #가중치
    bias = -0.6                            #치우침(편향)
    
    value = np.sum(input * weights) + bias
    
    if value <= 0:
        return 0
    else:
        return 1
    
```

```python
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

```python
x1 = np.arange(-2, 2, 0.01)
x2 = np.arange(-2, 2, 0.01)
bias = -0.6

y = (-0.4 * x1 - bias ) / 0.4

plt.axvline(x=0)          #수직축
plt.axhline(y=0)          #수평축

plt.plot(x1, y, 'r--')
plt.scatter(0,0,color='orange',marker='o',s=150)
plt.scatter(0,1,color='orange',marker='o',s=150)
plt.scatter(1,0,color='orange',marker='o',s=150)
plt.scatter(1,1,color='black',marker='^',s=150)
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
```

```python
def OR(x1, x2):
  input = np.array([x1, x2])
  weights = np.array([0.4,0.4])
  bias = -0.3

  value = np.sum(input * weights) + bias

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(OR(0,0))
print(OR(0,1))
print(OR(1,0))
print(OR(1,1))
```

```python
x1 = np.arange(-2, 2, 0.01)
x2 = np.arange(-2, 2, 0.01)
bias = -0.3

y = (-0.4 * x1 - bias ) / 0.4               #0.4 는 가중치

plt.axvline(x=0)          #수직축
plt.axhline(y=0)          #수평축

plt.plot(x1, y, 'r--')
plt.scatter(0,0,color='orange',marker='o',s=150)
plt.scatter(0,1,color='orange',marker='^',s=150)
plt.scatter(1,0,color='orange',marker='^',s=150)
plt.scatter(1,1,color='black',marker='^',s=150)
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
```

```python
def NAND(x1, x2):
  input = np.array([x1, x2])
  weights = np.array([-0.6,-0.6])               #가중치와 bias를 수정하면서 논리값을 만든다
  bias = 0.7

  value = np.sum(input * weights) + bias

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(NAND(0,0))
print(NAND(0,1))
print(NAND(1,0))
print(NAND(1,1))
```

```python
def XOR(x1, x2):                      #다층퍼셉트론을 구현하는 수식(인공지능학계의 잃어버린 10년을 깨는 순간)
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    
    return y
```

```python
print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
print(XOR(1,1))
```

```python
from keras import models
from keras import layers

network = models.Sequential()
network.add(layers.Dense(512, activation='relu', input_shape = (28 * 28,)))    #relu는 그래프를 찾아보라
network.add(layers.Dense(10, activation='softmax'))                    #softmax 는 총합은 1이고 나머지 요소들을 비율로 계산해줌
```

```python
network.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])           #rmsprop 가중치를 조절해주는 것
                                                                                                      #loss는 손실값 계산
                                                                                                      #metrics 지표 정확도
```

```python
train_images = train_images.reshape(60000, 28 * 28)
train_images.shape
```

```python
train_images = train_images.astype('float32') / 255
```

```python
train_images
```

```python
test_images = test_images.reshape(10000, 28*28)
test_images - test_images.astype('float32') / 255
```

```python
from tensorflow.keras.utils import to_categorical

train_labels = to_categorical(train_labels)                #to_categorical 원핫인코딩 = [ 10진법 -> 2진법 ]
test_labels = to_categorical(test_labels)
```

```python
network.fit(train_images, train_labels, epochs=5, batch_size=128)                       #accuracy : 훈련정확도
```

```python
test_loss, test_acc = network.evaluate(test_images, test_labels)            #loss 손실정도, #accuracy 테스트 정확도
```
